# **IMPORTING LIBRARIES**

```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# **LOADING DATASET**

```python
df= pd.read_excel('Windfarm.xlsx')
df.head()
```

```python
#Checking for null values
df.isnull().any()
```

```python
#Creating a copy of the dataframe
dfcpy=df.copy()
dfcpy
```

```python
#renaming a column name
dfcpy.rename(columns = {'Power (MW)':'Power'}, inplace = True)
dfcpy.rename(columns={'Time(year-month-day h:m:s)':'Time'},inplace = True)
dfcpy
```

# **SPLITTING INTO DEPENDENT AND INDEPENDENT VARIABLES**

```python
#dependent variable
y=dfcpy['Power']
y
```

```python
#independent variable
X=dfcpy.drop(columns=['Power','Time'],axis=1)
X.head()
```

# **SPLITTING THE DATASET INTO TRAINING AND TESTING**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)
#0.3 indicates 30% test dataset and remaining 70% training dataset which is ideal size of dataset for ml algorithms training and testing 
```

```python
X_train
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# **USING MULTI LINEAR REGRESSION MODEL**

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
y_pred
```

```python
Acc=pd.DataFrame({'Actual_y_value':y_test,'Predicted_y_value':y_pred})
Acc
```

# **VISUALIZATION OF MODEL'S PREDICTION**

```python
#Plotting the graph for test dataset 
plt.scatter(y_test, y_pred, color = 'red') 
 
plt.title( 'y_test vs y_pred (Test set)')
plt.xlabel('True values') 
plt.ylabel('Predictions') 
plt.show()
```
